# Ejercicio 2
La función, capitalizar_palabra, toma una palabra como entrada y devuelve la misma palabra con la primera letra en mayúscula y el resto en minúscula.

```python
def capitalizar_palabra( palabra: str):
    return palabra.capitalize()

capitalizar_palabra('prueba')
```

#### Crear un DataFrame de Pandas:
- Se crea un DataFrame llamado productos_df a partir de la lista de productos.

```python
productos_df = pd.DataFrame(productos, columns=["nombre", "precio", "cantidad_disponible"])
```

#### Calcular el total por producto y el valor total del inventario:
- Se calcula el total por producto multiplicando la cantidad de cada producto por su precio.
- Se crea una nueva columna "total_por_producto" en el DataFrame.
- Se calcula el valor total del inventario como la suma acumulada de los totales por producto.
- Se crea una nueva columna "valor_total_del_inventario" en el DataFrame.

```python
productos_df['total_por_producto'] = productos_df["precio"] * productos_df["cantidad_disponible"]
productos_df['valor_total_del_inventario'] = productos_df["total_por_producto"].cumsum()
```

#### Crear un nuevo DataFrame con el nombre y la cantidad disponible:
- Se crea un nuevo DataFrame llamado nombres_cantidad_df que contiene solo el nombre y la cantidad disponible de cada producto.

```python
nombres_cantidad_df = pd.DataFrame(productos_df[['nombre', 'cantidad_disponible']], columns=["nombre", "cantidad_disponible"])
```

# Ejercicio 4

#### Visualizar la cantidad disponible de productos:
- Se utiliza Matplotlib para graficar la cantidad disponible de cada producto.

```python
plt.bar(nombres_cantidad_df['nombre'], nombres_cantidad_df['cantidad_disponible'])
plt.title('Ejercicio 4')
plt.xlabel('Productos')
plt.ylabel('Cantidad')
plt.xticks(rotation=50, ha='right')
plt.show()
```

Este código crea un gráfico de barras que muestra la cantidad disponible de cada producto después de simular algunas ventas.
